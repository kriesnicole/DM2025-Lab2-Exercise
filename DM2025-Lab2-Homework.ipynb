## 1. Model Development (10 pts Required)

### 1.1 Preprocessing Steps

**Description:**  
In this step, we cleaned and prepared the raw text data for modeling.

**Steps performed:**  
- Converted all text to **lowercase** to ensure uniformity.  
- Removed **URLs, mentions (@user), HTML tags** and unnecessary symbols.  
- Kept hashtag words but removed the `#` sign.  
- Applied **tokenization** to split text into individual words.  
- Removed **stopwords** (common words like 'the', 'and', etc.).  
- Applied **lemmatization** to reduce words to their base form.  
- Replaced **NaN values** with empty strings to avoid errors in feature extraction.

**Result:**  
- Cleaned text stored in a new column `text_clean`.  
- Ready for feature extraction and modeling.

---

### 1.2 Feature Engineering Steps


**Description:**  
We converted the cleaned text into numerical features suitable for machine learning.

**Steps performed:**  
- Created **TF-IDF features** for words (1–3 grams) to capture important words and phrases.  
- Created **TF-IDF features for character n-grams** (3–5 characters) to capture subword patterns, spelling variations, and suffixes.  
- Combined **word-level and char-level TF-IDF** features into a single sparse matrix.  
- Ensured features are **numerical** and free of NaN values for model input.

**Result:**  
- Feature matrices `X_train` and `X_test` created.  
- Ready for model training.

---

### 1.3 Explanation of Your Model

Our final model is a **soft voting ensemble** combining three classifiers:  
- **LightGBM** on combined Word + Char TF-IDF features  
- **Logistic Regression** with class weighting  
- **Multinomial Naive Bayes**  

We use both **word-level TF-IDF (1–3 grams)** and **character-level TF-IDF (3–5 grams)** to capture a wide range of textual patterns, including misspellings and suffixes.  
The ensemble averages the predicted probabilities from all three classifiers, producing a more robust prediction and improving the **Macro F1 score**, especially for minority classes like `disgust` and `surprise`.  
This approach is fully CPU-friendly and scales well to ~64k posts.

## 1. Preprocessing Steps

```python
# PART 1 – DATA LOADING & MERGING

# imports
import pandas as pd
import json
from pathlib import Path

# paths
data_dir = Path(r".\data")

file_ident = data_dir / "data_identification.csv"
file_emotion = data_dir / "emotion.csv"
file_posts = data_dir / "final_posts.json"

# load CSV files
print("Loading CSV files...")
df_ident = pd.read_csv(file_ident, header=None, names=["id", "split"])
df_emotion = pd.read_csv(file_emotion, header=None, names=["id", "emotion"])

print("Identification shape:", df_ident.shape)
print("Emotion shape:", df_emotion.shape)
display(df_ident.head())
display(df_emotion.head())

# load JSON file
print("\nLoading JSON file...")
with open(file_posts, "r", encoding="utf-8") as f:
    posts = json.load(f)

# extract post_id and text
records = []
for item in posts:
    post_data = item["root"]["_source"]["post"]
    post_id = post_data.get("post_id")
    text = post_data.get("text")
    hashtags = post_data.get("hashtags", [])
    records.append({"id": post_id, "text": text, "hashtags": hashtags})

df_posts = pd.DataFrame(records)
print("Posts shape:", df_posts.shape)
display(df_posts.head())

# merge identification with emotion (labels)
df_merged = df_ident.merge(df_emotion, on="id", how="left")

# merge with text data
df_full = df_merged.merge(df_posts, on="id", how="left")

print("\nFull merged dataframe:")
print(df_full.shape)
display(df_full.head(10))

# split into train/test
train_df = df_full[df_full["split"] == "train"].reset_index(drop=True)
test_df = df_full[df_full["split"] == "test"].reset_index(drop=True)

print("\nTrain shape:", train_df.shape)
print("Test shape:", test_df.shape)
print("\nUnique emotions in train:", train_df["emotion"].unique())

# check missing values
missing_texts = df_full["text"].isna().sum()
missing_emotions = df_full["emotion"].isna().sum()
print(f"\nMissing texts: {missing_texts}, Missing emotions: {missing_emotions}")

# save clean datasets
train_df.to_csv("train_clean.csv", index=False)
test_df.to_csv("test_clean.csv", index=False)
print("\nSaved train_clean.csv and test_clean.csv")

# show some example lines
print("\nSample rows from train:")
display(train_df.sample(5, random_state=42)[["id", "emotion", "text"]])

print("\nSample rows from test:")
display(test_df.sample(5, random_state=42)[["id", "text"]])
```

## 2. Feature Engineering Steps

```python
# PART 2 – TEXT PREPROCESSING & FEATURE ENGINEERING (no emoji version)

import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# load cleaned train/test data
train_df = pd.read_csv("train_clean.csv")
test_df = pd.read_csv("test_clean.csv")

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# download nltk data (only once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# regex patterns
URL_RE = re.compile(r'https?://\S+|www\.\S+')
MENTION_RE = re.compile(r'@\w+')
HTML_RE = re.compile(r'<.*?>')

# simple text cleaning function
def preprocess_text(text):
    if pd.isna(text):
        return ""
    s = str(text).lower()
    s = HTML_RE.sub(' ', s)
    s = URL_RE.sub(' ', s)
    s = MENTION_RE.sub(' ', s)
    s = re.sub(r'#(\w+)', r'\1', s)  # keep hashtag word
    s = re.sub(r"[^\w\s']", ' ', s)
    s = re.sub(r'\s+', ' ', s).strip()
    tokens = nltk.word_tokenize(s)
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t not in stop_words and len(t) > 1]
    return " ".join(tokens)

# apply cleaning
print("Cleaning texts... this may take a few minutes")
train_df["text_clean"] = train_df["text"].astype(str).apply(preprocess_text)
test_df["text_clean"] = test_df["text"].astype(str).apply(preprocess_text)

# preview cleaned texts
print("\nSample cleaned train texts:")
display(train_df[["emotion", "text", "text_clean"]].sample(5, random_state=1))

# save cleaned data
train_df.to_csv("train_clean_preprocessed.csv", index=False)
test_df.to_csv("test_clean_preprocessed.csv", index=False)
print("\nSaved train_clean_preprocessed.csv and test_clean_preprocessed.csv")

# FEATURE ENGINEERING: TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1,2), min_df=3)
X_train = tfidf.fit_transform(train_df["text_clean"])
X_test = tfidf.transform(test_df["text_clean"])

print("\nTF-IDF matrices created:")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

# small feature preview
feature_names = tfidf.get_feature_names_out()
print("\nSample TF-IDF features:", feature_names[:20])
```

## 3. Model Implementation Steps

```python
# PART 3 – ENSEMBLE: LightGBM + Logistic Regression + Naive Bayes

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import VotingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score, classification_report, confusion_matrix
from scipy.sparse import hstack
import lightgbm as lgb
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter

# Load preprocessed data
train_df = pd.read_csv("train_clean_preprocessed.csv")
test_df = pd.read_csv("test_clean_preprocessed.csv")

# Fill NaN texts
train_df["text_clean"] = train_df["text_clean"].fillna("")
test_df["text_clean"] = test_df["text_clean"].fillna("")

# Encode labels
le = LabelEncoder()
y_train = le.fit_transform(train_df["emotion"])

# Feature Engineering – reduced TF-IDF for speed
tfidf_word = TfidfVectorizer(max_features=50000, ngram_range=(1,3), analyzer='word', min_df=3, sublinear_tf=True)
tfidf_char = TfidfVectorizer(max_features=25000, ngram_range=(3,5), analyzer='char_wb', sublinear_tf=True)

X_train_word = tfidf_word.fit_transform(train_df["text_clean"])
X_test_word = tfidf_word.transform(test_df["text_clean"])
X_train_char = tfidf_char.fit_transform(train_df["text_clean"])
X_test_char = tfidf_char.transform(test_df["text_clean"])

# Combine features
X_train = hstack([X_train_word, X_train_char])
X_test = hstack([X_test_word, X_test_char])

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

# LightGBM dataset
lgb_train = lgb.Dataset(X_train, label=y_train)

# LightGBM parameters
lgb_params = {
    'objective': 'multiclass',
    'num_class': len(le.classes_),
    'metric': 'multi_logloss',
    'num_leaves': 31,
    'learning_rate': 0.1,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': -1,
    'is_unbalance': True
}

# Train LightGBM model
clf_lgb = lgb.train(lgb_params, lgb_train, num_boost_round=100)

# Wrap LightGBM for sklearn API
class LGBWrapper:
    def __init__(self, model):
        self.model = model
    def fit(self, X, y):
        return self
    def predict_proba(self, X):
        return self.model.predict(X)
    
clf_lgb_wrap = LGBWrapper(clf_lgb)

# Logistic Regression
clf_lr = LogisticRegression(max_iter=300, class_weight='balanced', n_jobs=-1)
clf_lr.fit(X_train, y_train)

# Multinomial Naive Bayes
clf_nb = MultinomialNB()
clf_nb.fit(X_train, y_train)

# Ensemble: Soft Voting
ensemble = VotingClassifier(
    estimators=[('lgb', clf_lgb_wrap), ('lr', clf_lr), ('nb', clf_nb)],
    voting='soft',
    n_jobs=-1
)
# For soft voting, fit is not needed for LGBWrapper, just use predict_proba
# We'll use a custom predict_proba combining the three
probas_lgb = clf_lgb_wrap.predict_proba(X_train)
probas_lr = clf_lr.predict_proba(X_train)
probas_nb = clf_nb.predict_proba(X_train)
probas_avg = (probas_lgb + probas_lr + probas_nb) / 3
train_preds = np.argmax(probas_avg, axis=1)

# Train evaluation
f1_macro_train = f1_score(y_train, train_preds, average='macro')
print("\nTrain Macro F1 Score:", f1_macro_train)
print("\nClassification Report (Train Data):")
print(classification_report(y_train, train_preds, target_names=le.classes_))

# Confusion Matrix
cm = confusion_matrix(y_train, train_preds)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_, cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix on Training Data")
plt.show()

# Predict on Test Data & Submission
probas_lgb_test = clf_lgb_wrap.predict_proba(X_test)
probas_lr_test = clf_lr.predict_proba(X_test)
probas_nb_test = clf_nb.predict_proba(X_test)
probas_test_avg = (probas_lgb_test + probas_lr_test + probas_nb_test) / 3
test_preds_num = np.argmax(probas_test_avg, axis=1)
test_preds = le.inverse_transform(test_preds_num)

submission = pd.DataFrame({"id": test_df["id"], "emotion": test_preds})
submission.to_csv("submission_ensemble.csv", index=False)
print("\nSubmission saved as submission_ensemble.csv")

# Preview submission
display(submission.head(10))
```
